```python
import pandas as pd

dtypes = {
    "first_name": "category",
    "gender": "category",
    "type": "category",
    "state": "category",
    "party": "category",
}
df = pd.read_csv(
    "groupby-data/legislators-historical.csv",
    dtype=dtypes,
    usecols=list(dtypes) + ["birthday", "last_name"],
    parse_dates=["birthday"]
)

df
```

```python
# get warning if observed is not passed as keyword argument
state = df.groupby('state',sort=False)["last_name"].count()
state.tail()
```

```python
# suppress warning 
def decorator():   
    def Inner(func):       
        def wrapper(*args, **kwargs):
            #print('inner func',args[1],kwargs) 
            kwargs['observed']=False
            #print(args[0],args[1],kwargs)
            #func(*args, **kwargs)
            return args[0].groupby(args[1],**kwargs)["last_name"].count() 
        return wrapper
    return Inner
@decorator()
def grpby(*args, **kwargs):
    df_group = df.groupby(*args, **kwargs)["last_name"].count()
    #print(df_group)
grpby(df,"state", sort=False).tail()
```
