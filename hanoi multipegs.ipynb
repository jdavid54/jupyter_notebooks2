```python
def makehanoi(hn,debug=True):
    if not debug:print('-'.join([str(j) for j in hanois[-1]]),'(',hn,'pegs)')
    hanois.append([1])
    last = hanois[-1]
    previous = hanois[-2]
    
    for d in range(2,discs+1):
        if debug:
            print('\n',previous)
            print(last)
        #if debug:print(d,end='-')
        l = len(last)-1
        min = 2**d-1
        for k in range(d-l,d):
            l = d-k
            val = previous[k-1]+2*last[l-1]
            if val<=min:
                min = val
                sol = (d,k,l,min)
                if debug:print(sol,'+')
            else:
                sol = (d,k,l,val)
                if debug:print(sol)
        #if debug:print(min)
        last.append(min)
    
    return hanois
```

```python
discs = 10
n = list(range(1,discs+1))

#initialisation
previous = [2**k-1 for k in n]
hanois = [previous]

pegs = 5
for p in range(3,pegs+1):
    #print(hanois)
    hanois = makehanoi(p)
```
