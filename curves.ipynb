```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(-2,2,2000)
```

```python
y1=x**2
y2=x**3
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.grid(True)
plt.show()
```

```python
t=np.linspace(0,2*np.pi,2000)
```

```python
r=2
x=r*np.cos(t)
y=r*np.sin(t)
plt.plot(x,y,'r')
plt.axis('scaled')
plt.show()
```

```python
r=2*np.cos(2*t)
x=r*np.cos(t)
y=r*np.sin(t)
plt.plot(x,y)
plt.axis('scaled')
plt.show()
```

```python
r=2*np.cos(5*t)
x=r*np.cos(t)
y=r*np.sin(t)
plt.plot(x,y,'g')
plt.axis('scaled')
plt.show()
```
