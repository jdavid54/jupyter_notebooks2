```python
import numpy as np
import matplotlib.pyplot as plt


def binding_energy_per_nucleon(A, Z):
    # Empirical constants
    a_v = 15.75  # Volume term
    a_s = 17.8   # Surface term
    a_c = 0.711  # Coulomb term
    a_a = 23.7   # Asymmetry term

    # Pairing term
    if A % 2 == 1:  # Odd A
        delta = 0
    elif Z % 2 == 0:  # Even Z
        delta = 12 / np.sqrt(A)
    else:  # Odd Z
        delta = -12 / np.sqrt(A)

    # Binding energy
    E_b = (a_v * A
           - a_s * A**(2/3)
           - a_c * Z**2 / A**(1/3)
           - a_a * (A - 2 * Z)**2 / A
           + delta)

    return E_b / A

# Generate data for a range of nuclei
A_values = np.arange(2, 300)  # Mass numbers from 2 to 300
Z_values = A_values // 2      # Assume Z = A / 2 for simplicity (approximation for stable nuclei)

# Calculate average binding energy per nucleon
binding_energies = [binding_energy_per_nucleon(A, Z) for A, Z in zip(A_values, Z_values)]

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(A_values, binding_energies, label="Average Binding Energy")
plt.title("Average Binding Energy per Nucleon")
plt.xlabel("Mass Number (A)")
plt.ylabel("Binding Energy per Nucleon (MeV)")
plt.grid(True)
plt.legend()
plt.show()
```

### Explanation of `binding_energy_per_nucleon(A, Z)`

The `binding_energy_per_nucleon(A, Z)` function calculates the **average binding energy per nucleon** for a nucleus with a given mass number \( A \) (the total number of protons and neutrons) and atomic number \( Z \) (the number of protons). Below is a detailed explanation of the function:

#### **Function Parameters**
- **`A`**: The mass number (total number of nucleons, i.e., protons + neutrons).
- **`Z`**: The atomic number (number of protons).

#### **Constants**
The function uses the following empirical constants from the **semi-empirical mass formula (SEMF)**:
- **`a_v`**: Volume term coefficient. Represents the strong nuclear force binding nucleons together.
- **`a_s`**: Surface term coefficient. Accounts for nucleons at the surface having fewer neighbors to bind with.
- **`a_c`**: Coulomb term coefficient. Represents the repulsive force between protons due to their positive charge.
- **`a_a`**: Asymmetry term coefficient. Accounts for the energy cost of having unequal numbers of protons and neutrons.



#### **Pairing Term $$\delta$$**
The pairing term $$\delta $$ depends on whether the number of nucleons  A  is odd or even:

- $$\delta = 0 for odd A (nucleus has unpaired nucleons).$$

- $$\delta = +\frac{12}{\sqrt{A}}  for even  Z  (even-evennuclei   are   more stable).$$
- $$\delta = -\frac{12}{\sqrt{A}} for odd Z (odd-even nuclei are less stable).$$



#### **Binding Energy Calculation**
The total binding energy $$E_b$$ of the nucleus is calculated using the semi-empirical mass formula:

$$E_b = a_v A - a_s A^{2/3} - a_c \frac{Z^2}{A^{1/3}} - a_a \frac{(A - 2Z)^2}{A} + \delta$$



1. **Volume term represents the binding energy contributed by the strong nuclear force, proportional to A** $$a_v A$$
2. **Surface term Corrects for nucleons on the surface having fewer neighbors to bind with :** $$a_s A^{2/3} $$ 
3. **Coulomb term  Accounts for the electrostatic repulsion between protons.**
$$a_c \frac{Z^2}{A^{1/3}}$$
4. **Asymmetry term penalizes configurations where the neutron-to-proton ratio deviates significantly from stability.**$$a_a \frac{(A - 2Z)^2}{A}$$
5. **Pairing term Adds or subtracts energy based on the pairing of nucleons.**$$ \delta $$



#### **Average Binding Energy**
The function returns the **average binding energy per nucleon**:

$$
\text{Average Binding Energy per Nucleon} = \frac{E_b}{A}
$$

This value provides insight into how tightly each nucleon is bound within the nucleus.

#### **Use Case**
The function is useful for:
- Studying nuclear stability.
- Observing trends in nuclear binding energy across different nuclei.
- Understanding how certain configurations (e.g., even-even nuclei) are more stable.
- Exploring the variation of binding energy with the mass number \( A \).
