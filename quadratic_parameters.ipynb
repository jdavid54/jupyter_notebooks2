```python
from numpy import sqrt, linspace
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, b, c = 1, 0, 50
#b = sum of roots, can be written as b = (b/2a+u)+(b/2a-u)
#c = product of roots can be written as c = (b/2a+u)(b/2a-u) = b²/4a² - u²
# then  u² = b²/4a² - c
u2 = ((b/(2*a))**2 - c/a)
if u2 >= 0 : u = sqrt(u2)
else: u = 1j*sqrt(-u2)
c0 = (b/(2*a))**2*a
print(u2, u, c0)
```

```python
def f(x):
    return a*x**2+b*x+c
def locus(x):
    return -a*x**2+c

def solutions():
    u = (b/(2*a))**2-c/a
    if u >= 0:
        return (-b/(2*a)+sqrt(u), -b/(2*a)-sqrt(u))
    else:
        return (-b/(2*a)+1j*sqrt(-u), -b/(2*a)-1j*sqrt(-u))
```

```python
roots =(solutions())
r1, r2 = roots
print(r1, r2, r1+r2,r1*r2)
```

```python
def verify(roots):
    r1, r2 = roots
    return a*r1**2+b*r1+c,a*r2**2+b*r2+c
```

```python
verify(roots)
```
