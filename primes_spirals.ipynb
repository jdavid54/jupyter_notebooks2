```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def radial_plot(radius,theta,mks=5):
    plt.plot(radius*np.cos(theta),radius*np.sin(theta),'r',marker=".", markersize=mks,lw=0) 

def complex_plot(c ,mks=5):
    # extract real part 
    x = c.real
    # extract imaginary part 
    y = c.imag 
  
    # plotow() the complex numbers 
    plt.scatter(x, y, s=mks**2, c='b',lw=1) 
    plt.ylabel('Imaginary') 
    plt.xlabel('Real') 


def is_prime(n):
    if n % 2 == 0 and n > 2: 
        return False
    return all(n % i for i in range(3, int(np.sqrt(n)) + 1, 2))
```
