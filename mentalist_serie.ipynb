```python

def read_text(filename):
    f = open(filename, "r")
    text = f.readlines()
    f.close()
    print(text)


def get_data(filename):    
    data = pd.read_csv(filename)
    #data['time'] = pd.to_datetime(data['time'])  #.dt.tz_convert(None)
    #data.info()
    #print(data.tail(5))
    return data
```

```python
csv_file = r"C:\Users\jean\Desktop\Mentalist_s1_s7.csv"
df = get_data(csv_file)
```

```python

df[['Episodes', 'Titre']] = df['Titre'].str.split(" ", 1, expand=True)  # split 1 fois

df['Length']=df['Length'].str.extract(pat = '([0-9]*.[0-9]*)').astype('float')
df['Saisons']=df['Episodes'].str.extract(pat = '([S][0-9][0-7])')
df['Episodes']=df['Episodes'].str.extract(pat = '([E][0-9][0-9])')
df
```
