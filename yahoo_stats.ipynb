```python
presidents={
    'DeGaulle':[1959,1969],
    'Pompidou':[1969,1974],
    'Giscard_d_Estaing':[1974,1981],
    'Mitterrand':[1981,1995],
    'Chirac':[1995,2007],
    'Sarkozy':[2007,2012],
    'Hollande':[2012,2017],
    'Macron':[2017,2024]
}
```

```python
# Fonction pour obtenir le prÃ©sident correspondant Ã  une annÃ©e donnÃ©e
def get_president(year):
    for president, period in presidents.items():
        #print(president,period)
        if period[0] <= year < period[1]:
            return president
    return None
```

```python
# CrÃ©er une liste pour stocker les tuples (prÃ©sident, date)
index_list = []

# Remplir la liste avec les tuples correspondants
for date in ycac.index:
    president = get_president(date.year)
    index_list.append((president, date))
```

```python

# CrÃ©er un index hiÃ©rarchique
index = pd.MultiIndex.from_tuples(index_list, names=['PrÃ©sident', 'Date'])
index
```

```python

# RÃ©indexer le DataFrame avec l'index hiÃ©rarchique
ycac.index = index
```

```python
# Tracer les donnÃ©es par prÃ©sident
for president, data in ycac.groupby(level='PrÃ©sident'):
    plt.plot(data.index.get_level_values('Date'), data['Open'], label=president)

# Ajouter des lÃ©gendes et des titres
plt.legend()
plt.grid()
plt.title('CAC40 en fonction des PrÃ©sidents')
plt.xlabel('Date')
plt.ylabel('AnnÃ©e')
plt.show()
```
