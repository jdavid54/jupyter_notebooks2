```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sun Jul  7 20:58:53 2019
//////Get a image directory page and show all images/// 
@author: jeandavid
"""

import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (X11; Linux i686) AppleWebKit/537.36 (KHTML, like Gecko) Ubuntu Chromium/75.0.3770.90 Chrome/75.0.3770.90 Safari/537.36'}
URL='http://nhatbon.vn/upload/product/'
```

```python
#get the page
page=requests.get(URL,headers=headers)    
soup = BeautifulSoup(page.content, 'lxml')    
#search for <a> tag    
s=soup.find_all('a')
print(s)
```

```python
def scrap_image():
    #get the page
    page=requests.get(URL,headers=headers)    
    soup = BeautifulSoup(page.content, 'lxml')    
    #search for <a> tag    
    s=soup.find_all('a')
    text='''<!DOCTYPE HTML PUBLIC "-//W3C//DTD HTML 3.2 Final//EN">
    <html>
     <head>
        <meta content="text/html; charset=utf-8" http-equiv="Content-type"/>
        <title>Index of /upload/product/</title>
        <style>
            body {margin:0px;}
            .image {float: left;
                position: relative;
                width: 20%;
                height: 140px;
                border: 0px solid red;
            }
            img {width:100%;}
        </style> 
        </head>
        <body>
    '''
    #doing some tag arrangements
    n=0      #counting articles to show
    for i in s[5::2]:       # avoid no image links and duplicates 
        n+=1
        href=i['href'].replace('/upload','http://nhatbon.vn/upload')
        i.name='img'
        i['src']=href
        del i['href']
        i['title']=i.string
        i.string=''
        text+='<div class="image">'+str(i)+'</div>\n'        
    text+='</body></html>'
    text=text.replace('></img',' /')
    print(text)
    # create html file
    f = open("nhatbon.html", "w")
    f.write(text)
    f.close()
    return n
```

```python
n=scrap_image()
print(n,' articles to show in html page.')
```
