```python
import numpy as np
import matplotlib.pyplot as plt

def find_pythagorean_triplets(limit):
    triplets = []
    thetas = []
    for a in range(1, limit):
        for b in range(a, limit):
            c = (a**2 + b**2) ** 0.5            
            if c.is_integer() and c <= limit:
                triplets.append([a, b, int(c)])
                theta = np.arctan2(a,b)*180/np.pi
                #if theta not in thetas :
                thetas.append(theta)
    return triplets,thetas

limit = 50 #int(input("Enter the limit: "))
print("Pythagorean triplets up to", limit, "are:")
tr, th = find_pythagorean_triplets(limit)
print(tr)
```

```python

'''
TPP : triplet pythagore primitif
si (x ; y ; z) est un TPP, alors (y ; x ; z) est aussi un TPP, mais on choisit arbitrairement
de présenter le TPP avec x impair et y pair, z étant naturellement toujours impair.
Précisons que le terme y est en fait multiple de 4. en effet, si (x ; y ; z) est un tPP tel
que x = 2p + 1 et z = 2q + 1, p < q, alors :
y**2 = z**2 - x**2 = (2q + 1)**2 - (2p + 1)**2 = 4 [(q**2 + q) - (p**2 + p)]
Quelle que soit la parité de p et de q, les sommes p**2 + p et q**2 + q sont paires
et, par suite, la différence entre crochets est paire.
'''

def TPP_pythagore(a):
    '''
    Méthode attribuée à Pythagore (VIe siècle av. J.-C.)
    si a est un entier impair supérieur à 1, alors le triplet (x ; y ; z) tel que :
    x=a; y=1/2*(a**2-1); z=1/2*(a**2+1)
    est un TPP.
    '''
    b = 1/2*(a**2-1)
    c = 1/2*(a**2+1)
    if c.is_integer():
        return (a, b, c)

print('Méthode Pythagore')
for i in range(1,limit//2):
    try:
        x,y,z = TPP_pythagore(2*i+1)
        if x**2+y**2==z**2:
            print(x,int(y),int(z))
    except:
        pass
```

```python

def TPP_platon(k):
    '''
    Méthode attribuée à PLaton (IVe siècle av. J.C.)
    si k est un entier naturel non nul, alors le triplet (x ; y ; z) tel que
    x = 4k**2 - 1, y = 4k , z = 4k**2 + 1, est un TPP.
    Les cinq premières valeurs de k donnent (3 ; 4 ; 5), (15 ; 8 ; 17), (35 ; 12 ; 37),
    (63 ; 16 ; 65), (99 ; 20 ; 101). Ces tPP se caractérisent par l’égalité : z - x = 2.
    '''
    return (4*k**2-1, 4*k, 4*k**2+1)
    
print('Méthode Platon')    
for k in range(1,limit):   
    print(TPP_platon(k))
```

```python
    
    
def TPP_euclide(u):
    '''
    Théorème d’Euclide – aL KhÂZIn
    soit un couple d’entiers (u, v), non nuls, u > v, premiers entre eux et de parité différente.
    si on pose x = u**2 - v**2, y = 2uv, z = u**2 + v**2, alors (x ; y ; z) est un TPP.
    '''
    v = u-1
    return (u**2 - v**2, 2*u*v, u**2 + v**2)

print('Méthode Euclide')    
for k in range(2,limit):   
    print(TPP_euclide(k))
```

```python
def sqrt2_pythagorean_triplets(limit):
    triplets = []
    a, b, c = 3, 4, 5
    while c <= limit:
        triplets.append((a, b, c))
        new_a = 2 * b + c - a
        new_b = 2 * a + c - b
        new_c = 2 * a + 2 * b + c
        a, b, c = new_a, new_b, new_c
    return triplets

#limit = int(input("Entrez la limite : "))
print("Triplets pythagoriciens jusqu'à", limit, "sont :")
print(sqrt2_pythagorean_triplets(limit))
```

```python
def fibonacci_pythagorean_triplets(limit):
    triplets = []
    a, b = 1, 1
    while True:
        c = a + b
        if c > limit:
            break
        if (c - a) % 2 == 0:  # Ensure c - a is even
            triplets.append((int((c - a) / 2), a, int((c + a) / 2)))
        if (c - b) % 2 == 0:  # Ensure c - b is even
            triplets.append((int((c - b) / 2), b, int((c + b) / 2)))
        a, b = b, c
    return triplets

#limit = int(input("Entrez la limite : "))
print("Triplets pythagoriciens jusqu'à", limit, "sont :")
print(fibonacci_pythagorean_triplets(limit))
```

```python
def lucas_pythagorean_triplets(limit):
    triplets = []
    a, b = 2, 1
    while True:
        c = a + b
        if c > limit:
            break
        if (c - a) % 2 == 0:  # Ensure c - a is even
            triplets.append((int((c - a) / 2), a, int((c + a) / 2)))
        if (c - b) % 2 == 0:  # Ensure c - b is even
            triplets.append((int((c - b) / 2), b, int((c + b) / 2)))
        a, b = b, c
    return triplets

#limit = int(input("Entrez la limite : "))
print("Triplets pythagoriciens jusqu'à", limit, "sont :")
print(lucas_pythagorean_triplets(limit))
```

```python
def roberts_pythagorean_triplets(limit):
    triplets = []
    
    # Matrices de J. Roberts
    matrix1 = np.array([[1, -2, 2], [2, -1, 2], [2, -2, 3]])
    matrix2 = np.array([[1, 2, 2], [2, 1, 2], [2, 2, 3]])
    matrix3 = np.array([[-1, 2, 2], [-2, 1, 2], [-2, 2, 3]])

    matrices = [matrix1, matrix2, matrix3]

    for matrix in matrices:
        current_matrix = matrix.copy()
        while True:
            a = current_matrix[0, 0]
            b = current_matrix[1, 0]
            c = current_matrix[2, 0]

            if c > limit:
                break

            triplets.append((a, b, c))

            current_matrix = np.dot(matrix, current_matrix)

    return triplets

#limit = int(input("Entrez la limite : "))
print("Triplets pythagoriciens jusqu'à", limit, "sont :")
print(roberts_pythagorean_triplets(limit))
```

```python
def tchebychev_pythagorean_triplets(limit):
    triplets = []
    n = 1
    while True:
        m = n + 1
        while True:
            a = m**2 - n**2
            b = 2 * m * n
            c = m**2 + n**2
            k = 1
            while k * c <= limit:
                triplets.append((k * a, k * b, k * c))
                k += 1
            if a > b:
                break
            m += 1
        if 2 * n * (n + 1) > limit:
            break
        n += 1
    return triplets

#limit = int(input("Entrez la limite : "))
print("Triplets pythagoriciens jusqu'à", limit, "sont :")
print(tchebychev_pythagorean_triplets(limit))
```

```python
def conway_guy_pythagorean_triplets(limit):
    triplets = []
    matrix = np.array([[3, 4, -5], [4, 3, -5], [5, 12, -13]])
    current_matrix = matrix.copy()
    while True:
        a = current_matrix[0, 0]
        b = current_matrix[1, 0]
        c = current_matrix[2, 0]
        if c > limit:
            break
        triplets.append((a, b, c))
        current_matrix = np.dot(matrix, current_matrix)
    return triplets

#limit = int(input("Entrez la limite : "))
print("Triplets pythagoriciens jusqu'à", limit, "sont :")
print(conway_guy_pythagorean_triplets(limit))
```
