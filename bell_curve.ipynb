```python
import numpy as np
import matplotlib.pyplot as plt
sqrt = np.sqrt
pi = np.pi
exp = np.exp
```

```python
def f(x,u,s=1):
    return 1/(s*sqrt(2*pi))*exp(-1/2*((x-u)/s)**2)
```

```python
u = 0  # mean = central
n = 10  # number of curves
m = 3  # multiple of sigma s = width of the curves
# 68%(s=1), 95%(s=2), 99.7%(s=3)
for s in np.linspace(0.1,1,n):
    x = np.linspace(u-m*s,u+m*s,100)
    y = f(x,u,s)
    if s ==1:
        plt.plot(x,y,label='Standard (s=1)')
    else:
        plt.plot(x,y,label=str(s))
plt.legend()    
plt.grid()
```

```python
# rolling dices
import random
d = 10  # number of dices

def throw_dices():
    result = []
    for k in range(d):
        result.append(random.choice([1,2,3,4,5,6]))
    return result, sum(result)
```

```python
_r, _s = throw_dices()
_r, _s
```

```python
throws = 500
all_results = []

for k in range(throws):
    all_results.append(throw_dices()[1])
mu = np.mean(all_results)
mu
```
