```python
import requests
from bs4 import BeautifulSoup

# Make a request to the website
response = requests.get('https://coinmarketcap.com/fr/currencies/bitcoin/')
response.content
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
price_element = soup.find('div', class_='priceValue').children
l =[(k.name,k.contents[0]) for k in price_element]
l
```
