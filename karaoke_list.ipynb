```python
import csv
import pandas as pd

           
#read_csv(csv_file)
mypath = r'Z:\Videos\Films\____yggtorrent\Karaoké mp4'            
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
kara = pd.DataFrame(onlyfiles, columns =['Fichier'])
kara
```

```python
spl=kara['Fichier'].str.split(pat=' - ',expand=True)
spl.columns=['Chanteur', 'Chanson', 'ext1', 'ext2','ext3']
spl

```

```python
kara=pd.concat([kara, spl],axis=1)
kara
```

```python
kara1=kara[kara.ext1.isna()]
kara1
```

```python
kara2=kara[~kara.ext1.isna()]
#kara['Chanson']=kara[~kara.ext1.isna()]['ext1']
kara2
```

```python
kara2.loc[:,'Chanteur'] = kara2.loc[:,'Chanson']
kara2.loc[:,'Chanson'] = kara2.loc[:,'ext1']
kara2
```

```python
kara=pd.concat([kara1,kara2]).sort_index(ascending=True)
kara
```

```python
kara=kara.drop(columns=['ext1','ext2','ext3'])
kara
```

```python
spl2=kara['Chanson'].str.split(pat='.',expand=True,n=1)
spl2.columns=['Titre','extension']
spl2
```

```python
kara['Chanson']=spl2['Titre']
kara
```

```python
kara.to_csv('karaoke.csv',index=False)
```
