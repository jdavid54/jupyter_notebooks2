```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rotation(point, centre, theta):
    c, s = np.cos(theta), np.sin(theta)
    matrice = np.matrix([[c, -s],[s, c]])
    b2 = centre - np.dot(matrice, centre)
    #print(b2)
    rp = np.array(np.dot(matrice, point) + b2)[0]
    return rp
```

```python
p =(2,1)
c =(0,0.5)
rp = rotation(p,c,np.pi/3)
rp[0],rp[1]
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.axis([-5, 5, -5, 5])   #xmin, xmax, ymin, ymax

plt.plot(p[0],p[1],'bo')
plt.plot(c[0],c[1],'r*')
plt.plot(rp[0],rp[1],'go')
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
O =(0,0)

poly=[]
for k in range(20):   #len(vertexes)
    dots=[]
    p = vertexes[k]
    dots.append((p[0],p[1]))
    #print(p)
    #plt.plot(p[0],p[1],'ro')
    n = k + 3
    colors = ['ro','bo']*10
    # rotations successives du point vertex 
    for l in range(n):        
        rp = rotation(p,O,2*np.pi/n)
        p = (rp[0],rp[1])
        #print(p)
        plt.plot(rp[0],rp[1],colors[k])
        dots.append((rp[0],rp[1]))
    poly.append(dots)
```
