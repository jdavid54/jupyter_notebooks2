```python
import requests # https://requests.readthedocs.io/en/latest/user/quickstart/#make-a-request
from bs4 import BeautifulSoup

link = 'http://jean.david.free.fr/amlich/DuLieu/index_files/Sun-Moon-2020.html'

html_doc = requests.get(link)
html_doc.encoding = 'utf-8'

soup = BeautifulSoup(html_doc.text, 'html.parser')

heads = soup.find_all("td", class_="head")

year = '2025'
```

```python

for i, header in enumerate(heads):
    #print(header.get_text().strip())
    if header.get_text().strip() == year:
        print(f"\n=== Section: {header.get_text().strip()} ===")
        
        # Find all following siblings until next header
        next_sections = []
        current = header.find_next('tr')
        
        # search all td tags until td tag with class='head'
        while current and not current.find('td', class_='head'):
            next_sections.append(current)
            current = current.find_next('tr')
        
        # Print the data rows
        for row in next_sections:
            cells = row.find_all('td')
            if len(cells) == 3:  # Only process rows with 3 cells
                date1 = cells[0].get_text().strip()
                date2 = cells[1].get_text().strip()
                event = cells[2].get_text().strip()
                if date1 == '':
                    print(f"{date1} \t\t {date2} \t {event}")
                else:
                    print(f"{date1} \t {date2} \t {event}")
        break
```

```python

# Dictionary of all 24 solar terms with translations and significance
SOLAR_TERMS = {
    "Lập xuân": {
        "translation": "Start of Spring",
        "significance": "Marks the beginning of spring; agricultural preparations start"
    },
    "Vũ Thủy": {
        "translation": "Rain Water",
        "significance": "Increased rainfall, crucial for early crops"
    },
    "Kinh trập": {
        "translation": "Awakening of Insects",
        "significance": "Insects emerge from hibernation; farmers begin pest control"
    },
    "Xuân phân": {
        "translation": "Spring Equinox",
        "significance": "Equal day and night; planting season accelerates"
    },
    "Thanh minh": {
        "translation": "Pure Brightness",
        "significance": "Clear weather for tomb sweeping (Tết Thanh Minh) and planting"
    },
    "Cốc vũ": {
        "translation": "Grain Rain",
        "significance": "Rain nourishes young crops; ideal for sowing grains"
    },
    "Lập hạ": {
        "translation": "Start of Summer",
        "significance": "Transition to summer; rapid crop growth"
    },
    "Tiểu mãn": {
        "translation": "Grain Full",
        "significance": "Grains begin to plump but aren't fully ripe"
    },
    "Mang chủng": {
        "translation": "Grain in Ear",
        "significance": "Wheat/barley harvest; rice planting in southern regions"
    },
    "Hạ chí": {
        "translation": "Summer Solstice",
        "significance": "Longest day of the year; peak sunlight for crops"
    },
    "Tiểu thử": {
        "translation": "Minor Heat",
        "significance": "Hot weather begins; focus on irrigation"
    },
    "Đại thử": {
        "translation": "Major Heat",
        "significance": "Hottest period; drought risk management"
    },
    "Lập thu": {
        "translation": "Start of Autumn",
        "significance": "Transition to cooler weather; harvest preparations"
    },
    "Xử thử": {
        "translation": "End of Heat",
        "significance": "Heat subsides; ripening of late-summer crops"
    },
    "Bạch lộ": {
        "translation": "White Dew",
        "significance": "Morning dew appears; harvest season begins"
    },
    "Thu phân": {
        "translation": "Autumn Equinox",
        "significance": "Equal day and night; mid-autumn festivals (Tết Trung Thu)"
    },
    "Hàn lộ": {
        "translation": "Cold Dew",
        "significance": "Dew turns cold; rice harvesting in full swing"
    },
    "Sương giáng": {
        "translation": "Frost Descent",
        "significance": "First frost; protect tender crops"
    },
    "Lập đông": {
        "translation": "Start of Winter",
        "significance": "Winter begins; storage of harvested crops"
    },
    "Tiểu tuyết": {
        "translation": "Minor Snow",
        "significance": "Light snowfall; winter crop planting"
    },
    "Đại tuyết": {
        "translation": "Major Snow",
        "significance": "Heavy snow; livestock protection"
    },
    "Đông chí": {
        "translation": "Winter Solstice",
        "significance": "Shortest day; festivals celebrating the return of light"
    },
    "Tiểu hàn": {
        "translation": "Minor Cold",
        "significance": "Cold weather sets in; winter crops monitored"
    },
    "Đại hàn": {
        "translation": "Major Cold",
        "significance": "Year's coldest period; preparation for spring planting"
    }
}

def get_solar_terms(year='2039'):
    link = 'http://jean.david.free.fr/amlich/DuLieu/index_files/Sun-Moon-2020.html'
    
    try:
        # Fetch the HTML content
        html_doc = requests.get(link, timeout=10)
        html_doc.encoding = 'utf-8'
        
        # Parse with BeautifulSoup
        soup = BeautifulSoup(html_doc.text, 'html.parser')
        
        # Find all year headers
        heads = soup.find_all("td", class_="head")
        
        print(f"\nSolar Terms for Year {year}:")
        print("="*70)
        print(f"{'New Moon':<16}{'Term Date':<16}{'Vietnamese':<16}{'English':<20}{'Significance'}")
        print("-"*70)
        
        found = False
        for header in heads:
            if header.get_text().strip() == year:
                found = True
                # Get all rows until next year header
                current = header.find_next('tr')
                while current and not current.find('td', class_='head'):
                    cells = current.find_all('td')
                    if len(cells) == 3:
                        date1 = cells[0].get_text().strip()
                        date2 = cells[1].get_text().strip()
                        term = cells[2].get_text().strip()
                        
                        # Skip the header row
                        if term != "Tiết khí":
                            term_data = SOLAR_TERMS.get(term, {
                                "translation": "Unknown",
                                "significance": "Unknown term"
                            })
                            print(f"{date1 if date1 else '':<16}{date2:<16}{term:<16}{term_data['translation']:<20}{term_data['significance']}")
                    current = current.find_next('tr')
                break
        
        if not found:
            print(f"No data found for year {year}")
            
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")

# Example usage
get_solar_terms('2039')
```
