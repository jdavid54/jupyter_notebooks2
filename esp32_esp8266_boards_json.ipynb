```python


# URL of the JSON file
url = "https://raw.githubusercontent.com/espressif/arduino-esp32/gh-pages/package_esp32_index.json"

# Fetching the JSON data
response = requests.get(url)
data = response.json()
#type(data['packages'][0])
data
```

```python
# Extracting the 'packages' dictionary
packages_dict = data['packages'][0]
packages_dict
```

```python
df = pd.DataFrame.from_dict(packages_dict, orient='index')
df=df.transpose()
df
```

```python
# Creating an empty list to store the flattened dictionaries
flattened_data = []

# Flattening the dictionaries
#for package, package_info in platforms:
for package, package_info in platforms.items():
    print(package,":",package_info)
    if type(package_info) is dict:
        print(' is dict')
        for package2, package_info2 in package_info.items():
            flattened_info = pd.json_normalize(package_info)
            print(package2,":",package_info2)
            flattened_info[package2] = package_info2
            flattened_data.append(flattened_info)
    print()
    if type(package_info) is list:
        #print('is list')
        for _dict in package_info:
            print(_dict)
            for package2, package_info2 in _dict.items():
                flattened_info = pd.json_normalize(_dict)
                print(package2,":",package_info2)
                flattened_info[package2] = package_info2
                flattened_data.append(flattened_info)
```

```python
flattened_data
```

```python
# Concatenating the flattened dictionaries into a single DataFrame
df = pd.concat(flattened_data, ignore_index=True)
df
```
