```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2*np.pi,1000)
y = x**2
z = 2*x  #dérivée de y
```

```python
t = np.linspace(0,10,1000)
a, y0 = 1, 100
s = a
y = t*np.exp(a*t) + y0*np.exp(a*t)
dy = np.exp(a*t) + a*t*np.exp(a*t) + y0*np.exp(a*t)
dy2 = a*y + np.exp(s*t)
#plt.plot(t,y)
plt.plot(t,dy)
plt.plot(t,dy2)
plt.show()

assert dy.any()==dy2.any()
```

```python
s = a
if s == a :
    y = t*np.exp(a*t) + y0*np.exp(a*t)
    dy2 = np.exp(a*t) + a*t*np.exp(a*t) + y0*np.exp(a*t)
else:
    y = (np.exp(s*t) - np.exp(a*t)) / (s - a) + y0*np.exp(a*t)
    dy2 = (s*np.exp(s*t) - a*np.exp(a*t)) / (s - a) + a*y0*np.exp(a*t)

dy = a*y + np.exp(s*t)
#plt.plot(t,y)
plt.plot(t,dy,label='dy')
plt.plot(t,dy2,label='dy2')
plt.legend()
plt.show()

assert dy.any()==dy2.any()
```

```python
# dy = ay + cos(wt)  (0)
a, w = 1, 1
M = a/(w**2 + a**2)    # G > 0 then M > 0
N  = -w/(w**2 + a**2)
print(M, N)

# (1) première forme solution particulière
y1 = M*np.cos(w*t) + N*np.sin(w*t)  # (1)

#(2) 2eme forme de solution particulière : y = G.cos(wt - alpha)
G = np.sqrt(M**2 + N**2)  # G > 0
alpha = np.arctan(N/M)
print(G, alpha)
print(M,G*np.cos(alpha))
print(N,G*np.sin(alpha))

y2 = G*np.cos(w*t - alpha)  # (2)

plt.plot(t,y1,label='y1')
plt.plot(t, y2,label='y2') 
plt.legend()
plt.show()

assert y1.any()==y2.any()
```

```python
dy1 = a*y1 + np.cos(w*t)        # (0)
dy2 = a*y2 + np.cos(w*t)
plt.plot(t,dy1, label='dy1')
plt.plot(t, dy2, label='dy2')   # dy1 = dy2
plt.legend()
plt.show()

assert dy1.any()==dy2.any()
```

```python
dy1 = -w*M*np.sin(w*t) + w*N*np.cos(w*t)
dy2 = -w*G*np.sin(w*t - alpha)
plt.plot(t,dy1, label='dy1')
plt.plot(t, dy2, label='dy2')   # dy1 = dy2
plt.legend()
plt.show()

assert dy1.any()==dy2.any()
```
