```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dim = 10000
#bins = 200
```

```python
# random uniform variable
x = np.random.random(dim)
x = np.random.ranf(dim)
x = np.random.sample(dim)
count, bins, ignored =plt.hist(x, 200, density=True) 
avrg = sum(count)/200
plt.grid()
plt.plot(bins, avrg*np.ones_like(bins), linewidth=2, color='r')
plt.title('Random series')
plt.show()
```

```python
s = np.random.uniform(-1,0,dim)
print(s)
isgreaterlow=np.all(s >= -1)
islowerhigh=np.all(s < 0)
print(isgreaterlow,islowerhigh)
count, bins, ignored = plt.hist(s, 200, density=True)
#print(bins)
#print(count)
#print(np.ones_like(bins))
avrg = sum(count)/200
#plt.plot(bins[:200],count[:200])
plt.plot(bins, avrg*np.ones_like(bins), linewidth=2, color='r')
plt.title('Random uniform')
plt.grid()
plt.show()
print(avrg)
```

```python
#triangular distribution
t = np.random.triangular(-1, 0, 1, dim)
count, bins, ignored = plt.hist(t, bins=200, density=True)
#print(h)
avrg = sum(count)/200
plt.plot(bins, avrg*np.ones_like(bins), linewidth=2, color='r')
plt.title('Random triangular')
plt.grid()
plt.show()
print(avrg)
```

```python
#Theorem The difference of two independent standard uniform random
#variables has the standard trianglular distribution.
x1 = np.random.uniform(0,1,dim)
x2 = np.random.uniform(0,1,dim)
x = x1 - x2
plt.hist(x, 200, density=True)
#plt.show()

y1 = np.random.uniform(0,1,dim)
y2 = np.random.uniform(0,1,dim)
y = y1 - y2
plt.hist(y, 200, density=True)
plt.title('Difference of random uniform series')
plt.grid()
plt.show()
```

```python
x = x1**2
count, bins, ignored = plt.hist(x, 200, density=True)
avrg = sum(count)/200
plt.plot(bins, avrg*np.ones_like(bins), linewidth=2, color='r')
plt.title('Square of random uniform series')
plt.grid()
plt.show()
```

```python
x = np.sqrt(x1)
plt.hist(x1, 200, density=True)
plt.grid()
plt.show()
count, bins, ignored = plt.hist(x, 200, density=True)
avrg = sum(count)/200
plt.plot(bins, avrg*np.ones_like(bins), linewidth=2, color='r')
plt.title('Square root of random uniform series')
plt.grid()
plt.show()
```

```python
x1 = np.random.uniform(0,1,dim)
x2 = np.random.uniform(0,1,dim)

plt.hist(x1, 200, density=True)
plt.hist(x2, 200, density=True)
plt.grid()
plt.show()
```

```python
x = (x1 + x2)
count, bins, ignored = plt.hist(x, 200, density=True)
avrg = sum(count)/200
plt.plot(bins, avrg*np.ones_like(bins), linewidth=2, color='r')
plt.title('Sum of random uniform series')
plt.grid()
plt.show()
```

```python
x = x1*x2
count, bins, ignored = plt.hist(x, 200, density=True)
avrg = sum(count)/200
plt.plot(bins, avrg*np.ones_like(bins), linewidth=2, color='r')
plt.title('Product of random uniform series')
plt.grid()
plt.show()
```
