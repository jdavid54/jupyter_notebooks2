```python
import random
```

```python
def walk(n):
    '''distance after n random steps, compact code'''
    x,y=0,0
    for i in range(n):
        (dx,dy)=random.choice([(0,1),(0,-1),(1,0),(-1,0)])    #N,S,E,W
        x+=dx
        y+=dy
    return x,y
```

```python
def minmaxdots(n):           # n=number of walks
    '''for each walk length, average over n walks in % when distance <= 4 (ie no-transport needed)'''
    x1=([])
    y1=([])
    x2=([])
    y2=([])
    n1=0
    n2=0
    walk_length=maxlength
    for i in range(n):
        x,y=walk(walk_length)
        distance=abs(x)+abs(y)
        if distance>max:        #distance max
            x2.append(x)
            y2.append(y)
            n2+=1
            #print('x',end='')
        else:
            x1.append(x)
            y1.append(y)
            n1+=1
            #print('o',end='')
    return x1,y1,x2,y2,n1,n2
```

```python
maxlength=22
max=4
iteration=5000
x1,y1,x2,y2,n1,n2=minmaxdots(iteration)
print('iteration=',iteration,'inbound=',n1,'outbound=',n2,'percentage_inbound=',f'{100*n1/iteration:.2f}')           
```
