```python
csv_file = 'titanic.csv'
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

with open(csv_file) as sourcefile:
    df=pd.read_csv(sourcefile, delimiter=",")
```

```python
df
```

```python

fig, ax = plt.subplots()
size = 0.4

underage = df[df.Age<18]
youngadult = df[df. Age>=18][df.Age<30]
adult = df[df. Age>=30][df.Age<65]
elderly = df[df.Age>=65]

age_div = [len(underage.index),len (youngadult.index),len (adult.index),len (elderly.index)]
age_div = np.array(age_div)

underage_nonalive = df[df.Age<18][df.Survived==0]
youngadult_nonalive = df[df.Age>=18][df. Age<30][df.Survived ==0]
adult_nonalive = df[df. Age>=30][df.Age<65][df.Survived==0]
elderly_nonalive = df[df.Age>=65][df.Survived==0]

age_div_survivors = [len(underage_nonalive.index),len(youngadult_nonalive.index),len(adult_nonalive.index),len (elderly_nonalive.index)]
age_div_survivors = np.array(age_div_survivors)

cmap = plt.colormaps ["tab20c"]
outer_colors = cmap([11, 10, 9, 8])
inner_colors = cmap([21, 18, 17, 16])
labels = ["Children/under 18", "Young adults", "Adults", "Elderly"]
ax.pie(age_div.flatten(), radius=1, colors=outer_colors, labels=labels,
            wedgeprops=dict(width=size, edgecolor='w'))

ax.pie(age_div_survivors.flatten(), radius=1-size, colors=inner_colors, wedgeprops=dict(width=size, edgecolor='w'))

ax.set(aspect="equal", title='Comparison of passenger % in age groups vs non-surviver % in age groups')

#plt.show()
```
