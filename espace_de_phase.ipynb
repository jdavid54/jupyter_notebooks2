```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
w = 1
x = np.linspace(0,2*np.pi,1000)
y = np.sin(w*x)
z = w*np.cos(w*x)  #dérivée de y
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
#espace de phase (y,y')
plt.plot(y,z)
plt.show()
```

```python
#espace de phase (y,y')
plt.plot(z,-y)
plt.show()
```
