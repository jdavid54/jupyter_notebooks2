```python
import numpy as np

def trapezes(f,a,b,n):
    t, h, s = a, (b-a)/n, (f(a)+f(b))/2
    for k in range(1,n):
        t = t+h
        s = s+f(t)
    return s*h

trapezes(np.sin,0,np.pi,100)
```

```python
trapezes(np.sin,0,np.pi,1000)
```

```python
trapezes(np.cos,0,np.pi,1000)
```

```python
realpart=[]
imgpart =[]

def trapezesEchant(Y, a, b):
    n = len(Y)-1
    t, h, s = a, (b-a)/n, (Y[0]+Y[-1])/2
    for k in range(1, n-1):
        s = s + Y[k]
        #print(s)
        realpart.append(np.real(s))
        imgpart.append(np.imag(s))
    return s*h
```

```python
w = 1
G = lambda x : np.exp(1j*w*x)

X = np.linspace(0, np.pi, 1001)
Y = np.sin(X)
Z = Y*G(X)

R1 = trapezesEchant(Z, 0, np.pi)
R1
```

```python
import matplotlib.pyplot as plt
print(max(realpart))
plt.plot(realpart,imgpart)
plt.show()
```
